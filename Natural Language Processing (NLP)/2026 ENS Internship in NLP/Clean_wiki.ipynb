## ---- Create cleaned wiki data ----

We want to :
- split every articles by /n and then choose a random paragraph with >= 10 words
- go from 100k to 100 random articles

#### 1. Import Librairies 

```python
import pandas as pd
import random
```

#### 2. Split /n and choose a random paragraph

```python
def process_articles(csv_path, text_column="text", min_words=10, random_state=None):

    if random_state is not None:
        random.seed(random_state)

    df = pd.read_csv(csv_path)

    cleaned_texts = []

    for article in df[text_column].astype(str):
        # Split into paragraphs
        paragraphs = [p.strip() for p in article.split("\n") if p.strip()]

        # remove short paragraphs
        long_paragraphs = [
            p for p in paragraphs
            if len(p.split()) >= min_words
        ]

        if not long_paragraphs:
            cleaned_texts.append(None)
        else:
            # Keep one paragraph randomly
            chosen = random.choice(long_paragraphs)
            cleaned_texts.append(chosen)

    # Replace original column
    df[text_column] = cleaned_texts

    # Drop rows where no paragraph survived
    df = df.dropna(subset=[text_column])

    return df
```

```python
# Do it on our data
if __name__ == "__main__":
    input_path = "wiki_en_20231101_100k_articles.csv"
    df_clean = process_articles(input_path, text_column="text", min_words=10, random_state=42)
    df_clean.to_csv("wiki_cleaned_one_paragraph_per_article_100k.csv", index=False)
    print("Done.")
```
